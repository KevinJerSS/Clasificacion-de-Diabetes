## 1. Definición del problema

> Dados los parámetros clínicos de un paciente, ¿podemos predecir si tiene o no diabetes?

## 2. Datos

Este conjunto de datos proviene originalmente del Instituto Nacional de Diabetes y Enfermedades Digestivas y Renales. El objetivo es predecir, a partir de mediciones diagnósticas, si un paciente tiene diabetes.

Disponible versión en Kaggle. https://www.kaggle.com/datasets/pavansanagapati/pima-diabetes?select=pimaindians-diabetes.data.csv

## 3. Evaluación

> Si podemos alcanzar una precisión del 95% al predecir si un paciente tiene o no diabetes durante la prueba de concepto, seguiremos adelante con el proyecto.

## 4. Características

Aquí obtendrá información diferente sobre cada una de las características de sus datos. Puede hacerlo investigando por su cuenta (por ejemplo, consultando el enlace anterior) o hablando con un experto en la materia (alguien que conozca el conjunto de datos).

**Crear diccionario de datos**
1. Embarazos: Número de embarazos
2. Glucosa: Concentración plasmática de glucosa a las 2 horas en una prueba de tolerancia oral a la glucosa
3. Presión arterial: Presión arterial diastólica (mmHg)
4. Grosor de la piel: Grosor del pliegue cutáneo tricipital (mm)
5. Insulina: Insulina sérica a las 2 horas (μU/ml)
6. IMC: Índice de masa corporal (peso en kg/(altura en m)^2)
7. Función genealógica de la diabetes: Función genealógica de la diabetes
8. Edad: Edad (años)
9. Resultado: Variable de clase (0 = NO o 1 = SI)

## Preparando las herramientas
Utilizaremos pandas, Matplotlib y NumPy para el análisis y manipulación de datos.

```python
# Importar todas las herramientas que necesitamos 
# Bibliotecas de gráficos y análisis exploratorio de datos 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib import rcParams
sns.set()

# Clasificación multiclase
from pycaret.classification import *

# Modelos de Scikit-Learn y Evaluación de modelos
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.metrics import f1_score, confusion_matrix, precision_recall_curve, roc_curve
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings(action='ignore')
```

```python
def get_clf_eval(y_test, pred=None, pred_proba=None):
    confusion = confusion_matrix( y_test, pred)
    accuracy = accuracy_score(y_test , pred)
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    f1 = f1_score(y_test,pred)   
     
    roc_auc = roc_auc_score(y_test, pred_proba)

    # ROC-AUC print 
    print('accuracy: {0:.4f}, precision: {1:.4f}, recall: {2:.4f},\
    F1: {3:.4f}, AUC:{4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
    return confusion
```

Por lo tanto, decidimos realizar un escalado no lineal y utilizar el Transformador Cuantil, que modifica la distribución más cercana a la distribución normal, consultando el cuaderno mencionado a continuación.

La función cuantil jerarquiza o suaviza la relación entre las observaciones y puede aplicarse a otras distribuciones, como la distribución uniforme o la normal.

```python
from sklearn.preprocessing import QuantileTransformer
scaler = QuantileTransformer(n_quantiles=100, random_state=0, output_distribution='normal')
X_scaled = scaler.fit_transform(X)
```

### Top 5 de metodos de clasificación

#### Setup 

Esta función inicializa el entorno de entrenamiento y crea la canalización de transformación. Debe llamarse antes de ejecutar cualquier otra función. Acepta dos parámetros obligatorios: datos y destino. Los demás parámetros son opcionales.

```python
# Inicialización del entorno de PyCaret
setup(df, target='Resultado', preprocess=False)
```

#### compare_models

Esta función entrena y evalúa el rendimiento de todos los estimadores disponibles en la biblioteca de modelos mediante validación cruzada. El resultado de esta función es una cuadrícula de puntuaciones con las puntuaciones promedio de la validación cruzada.

```python
top_5 = compare_models(sort='AUC',
               n_select=5,
               exclude=['lightgbm','xgboost','dummy','svm','ridge','knn','dt','nb','qda'])
```

#### Creating Models

Esta función entrena y evalúa el rendimiento de un estimador determinado mediante validación cruzada. El resultado es una cuadrícula de puntuaciones con puntuaciones de CV por pliegue.

```python
# Linear Discriminant Analysis
lda = create_model('lda')

# Logistic Regression
lr = create_model('lr')

# Extra Trees Classifier
et = create_model('et')

# Random Forest Classifier
rf = create_model('rf')

# Gradient Boosting Classifier
gbc	= create_model('gbc')
```

### Modelo de interpretación

```python
interpret_model(rf)
```

```python
interpret_model(et)
```

#### Observación: 
- Como era de esperar, la glucosa se considera la característica más importante. 
- El Grosor de piel y la presión arterial tienen menor importancia.

### Ajuste de hiperparámetros

Esta función ajusta los hiperparámetros de un estimador dado. El resultado es una cuadrícula de puntuaciones con puntuaciones de CV por pliegue del mejor modelo seleccionado, según el parámetro de optimización.

```python
# Linear Discriminant Analysis
tuned_lda = tune_model(lda, optimize = 'AUC')

# Logistic Regression
tuned_lr = tune_model(lr, optimize = 'AUC')

# Extra Trees Classifier
tuned_et = tune_model(et, optimize = 'AUC')

# Random Forest Classifier
tuned_rf = tune_model(rf, optimize = 'AUC')

# Gradient Boosting Classifier
tuned_gbc = tune_model(gbc, optimize = 'AUC')
```

Los modelos Top 1, Top 2, Top 3, Top 4 y Top 5 se ajustaron con diferente importancia de característica y límite de decisión. Además, existe una gran diferencia en la importancia de característica con respecto al clasificador catboost.

### Stacking

Teniendo nuestro top 5 mejores modelos, usamos el primer modelo para que aprenda las predicciones de los demás.

```python
stack_model = stack_models(estimator_list=top_5, 
                           meta_model=top_5[0],
                           optimize='AUC')
```

```python
plt.figure(figsize=(8, 8))
plot_model(stack_model, plot='boundary')
```

```python
# Grafico la curva ROC y calculo la métrica AUC
plt.figure(figsize=(8, 8))
plot_model(stack_model, plot = 'auc')
```

```python
# Predicción
pred = stack_model.predict(X_test)
pred_proba = stack_model.predict_proba(X_test)[:,1]

# Accuracy
confusion_stack = get_clf_eval(y_test,pred,pred_proba)
```

```python
# Matriz de confusión
plt.figure(figsize=(8, 6))
ax = sns.heatmap(confusion_stack, cmap='YlGnBu', annot=True, fmt='d')
ax.set_title('Matriz de confusión (Stacking)');
```

La precisión y la recuperación tienen una relación de equilibrio. La imagen de arriba parece estar bien equilibrada porque ha sido bien estudiada.

### Soft

Promediar las probabilidades, usando un enfoque más simple que stacking

```python
blend_soft = blend_models(estimator_list=top_5, optimize='AUC', method='soft')
```

```python
plt.figure(figsize=(8, 8))
plot_model(blend_soft, plot='boundary')
```

Parece que el modelo ha sido entrenado correctamente.

```python
# Grafico la curva ROC y calculo la métrica AUC
plt.figure(figsize=(8, 8))
plot_model(blend_soft, plot = 'auc')
```

```python
# Predección
pred = blend_soft.predict(X_test)
pred_proba = blend_soft.predict_proba(X_test)[:,1]
# Accuracy
confusion_soft = get_clf_eval(y_test,pred,pred_proba)
```

```python
# Matriz de confusión
plt.figure(figsize=(8, 6))
ax = sns.heatmap(confusion_soft, cmap='YlGnBu', annot=True, fmt='d')
ax.set_title('Matriz de confusión (Soft Blending)');
```
